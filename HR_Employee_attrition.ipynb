## Data preprocessing

```python
#checking if dataframe is having any null value

df.isnull().sum()
```

```python
# checking for duplicated rows

df.duplicated().sum()
```

```python
df.info()
```

```python
#separating categorical and numerical features

catvars = df.select_dtypes(include=['object']).columns
numvars = df.select_dtypes(include = ['int32','int64','float32','float64']).columns

catvars,numvars
```

```python
#knowing unique values of all the columns
def uniquevals(col):
    print(f'Details of the particular col {col} is : {df[col].unique()}')
    
def valuecounts(col):
    print(f'Valuecounts of the particular col {col} is : {df[col].value_counts()}')
    
    
    
    
for col in df.columns:
    uniquevals(col)
```

#### The columns having only one unique value are of no use as they are same for every element and hence does not affect the attrition at all

```python
df=df.drop(['EmployeeNumber','EmployeeCount','Over18','StandardHours'],axis=1)
df
```

```python
#replacing Yes with 1 and no with 0
df['Attrition'] = df['Attrition'].apply(lambda element:1 
                                                      if 'Yes' in element else 0)

df.head()
```

```python
df.dtypes['Attrition']
```

```python
#replacing 1=Travel Rarely and 2=Travel_Frequently and 0=Non travel
df['BusinessTravel'] = df['BusinessTravel'].apply(lambda element:1 
                                                  if 'Travel_Rarely' in element else(2 if 'Travel_Frequently' in element else 0))
df.head()
```

```python
#replacing 1=Sale and 2=Research & Development and 0=others
df['Department'] = df['Department'].apply(lambda element:1 
                                                  if 'Sales	' in element else(2 if 'Research & Development' in element else 0))
df.head()
```

```python
#replacing Male with 1 and female with 0
df['Gender'] = df['Gender'].apply(lambda element:1 
                                    if 'Male' in element else 0)

df['Gender']
```

```python
#replacing 1=Single and 2=Married and 0=Divorced
df['MaritalStatus'] = df['MaritalStatus'].apply(lambda element:1 
                                                  if 'Single' in element else(2 if 'Married' in element else 0))
df['MaritalStatus']
```

```python
#replacing Yes with 1 and no with 0
df['OverTime']= df['OverTime'].apply(lambda element:1 
                                        if 'Yes' in element else 0)
df['OverTime']
```

```python
df
```

```python
df.corr()['Attrition']
```

#### There are negative correlation values which indicates negative correlation i.e when that feature increases attrition decreases

```python
plt.figure(figsize=(15,20))
sn.heatmap(df.corr(),annot=True,cmap='summer',fmt='.0%')
```

```python
#converting the remaining catvars colums into numvars columns
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
df
```

### Feature Selection

```python
x = df.drop(columns = "Attrition", axis = 1)
y = df["Attrition"]
```

### MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
random_model=RandomForestClassifier(n_estimators = 250)
```

#### Fit

```python
random_model.fit(xtrain , ytrain)
```

#### Checking the Accuracy

```python
#Training Accuracy
random_model_accuracy = round(random_model.score(xtrain,ytrain)*100,2)
print(round(random_model_accuracy,2) , '%')
```

```python
#testing Accuracy
random_model_accuracy1= round(random_model.score(xtest,ytest)*100,2)
print(round(random_model_accuracy1,2) , '%')
```
